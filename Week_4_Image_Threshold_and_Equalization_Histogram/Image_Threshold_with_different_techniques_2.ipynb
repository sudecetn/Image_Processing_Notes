```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
my_gray_img = cv2.imread("./kapadokya.jpg",0)
my_gray_img.shape
```

```python
_, my_binary_thresh_img = cv2.threshold(my_gray_img, thresh=128, maxval=255,type=cv2.THRESH_BINARY)
```

```python
fig = plt.figure(figsize=(10,7))
rows = 1
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_gray_img, cmap= "gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_binary_thresh_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRES IMAGE")
```

```python
_, my_binary_inv_thresh_img = cv2.threshold(my_gray_img, thresh =128, maxval=255,type=cv2.THRESH_BINARY_INV)
```

```python
fig = plt.figure(figsize=(14,7))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(my_gray_img, cmap= "gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_binary_thresh_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRES IMAGE")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_binary_inv_thresh_img, cmap="gray")
plt.axis('off')
plt.title("BINARY INV THRES IMAGE")
```

```python
_, my_thresh_trunc = cv2.threshold(my_gray_img, thresh =128, maxval=255,type=cv2.THRESH_TRUNC)
_, my_thresh_tozero = cv2.threshold(my_gray_img, thresh =128, maxval=255,type=cv2.THRESH_TOZERO)
_, my_thresh_tozero_inv = cv2.threshold(my_gray_img, thresh =128, maxval=255,type=cv2.THRESH_TOZERO_INV)
```

```python
fig = plt.figure(figsize=(15,7))
rows = 2
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(my_gray_img, cmap= "gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_binary_thresh_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDİNG IMAGE")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_binary_inv_thresh_img, cmap="gray")
plt.axis('off')
plt.title("BINARY INV THRESHOLDİNG IMAGE")

fig.add_subplot(rows, columns, 4)
plt.imshow(my_thresh_trunc, cmap="gray")
plt.axis('off')
plt.title("TRUNC THRESHOLDİNG IMAGE")

fig.add_subplot(rows, columns, 5)
plt.imshow(my_thresh_tozero, cmap="gray")
plt.axis('off')
plt.title("TOZERO THRESHOLDİNG IMAGE")

fig.add_subplot(rows, columns, 6)
plt.imshow(my_thresh_tozero_inv, cmap="gray")
plt.axis('off')
plt.title("TOZERO INV THRESHOLDİNG IMAGE")
```
